```python
import pandas as pd
import numpy as np
```

$$ k = \frac{A_{pGRR} - \beta G}{\beta G r^{n}} $$
$$ r = \frac{A_{ADH1}}{b} $$

Unknowns:

    RGR expression level from ADH1
    Promoter strength of pGRR (estimated from "normalizing_proms"). 
$$ AU * \beta = A_{pGRR} $$
   

```python
ADH1 = 4.0
b = 0.11997332
n = 1.630441275
r = ADH1/b
```

```python
df = pd.read_csv("/Users/soslab/Downloads/20150209_RealFullWxFset.csv")
```

```python
df.columns
```

```python
df = pd.DataFrame({'proms': df['proms'], 
                  'RGRs': df['RGRs'],
                  'pGRR_norm': df['normalizing_proms']/1600.,
                  'rel_exp': df['expression'],
                  'AU': df['FL1.Amean']/1600.0}
                 )
```

```python
df['A_pGRR'] = df['pGRR_norm']*b
df['k'] = (df['A_pGRR']-b*df['AU'])/(b*df['AU']*r**n)
df
```
